## Théorème des écarts complémentaires
Soit x et λ des solutions pour les programmes primal et dual, le
primal étant exprimé sous forme standard. Une condition nécessaire et suffisante pour que x et λ soient tous deux solutions optimales est que pour tout i

1) xi > 0 ⇒ λ
T ai = ci
2) xi = 0 ⇐ λ
T ai < ci



On utilise le théorème des écarts complémentaires pour trouver sa solution optimale :  


$$
\begin{cases}
x + 2y &= 2 \\
2x - y &= 3
\end{cases}
$$

$$⇒
\begin{cases}
x + 2(2x-3) &= 2 \\
y &= - 3 + 2x
\end{cases}
$$

$$⇒
\begin{cases}
x + 4y - 6 &= 2 \\
y &= - 3 + 2x
\end{cases}
$$

$$⇒
\begin{cases}
x &= 8/5 \\
y &= 1/5
\end{cases}
$$

Ainsi la solution optimale du dual du problème est (8/5, 1/5)

```julia
# Fonction pour afficher les tableaux de chaque itération
function displayIteration(M::Matrix, iteration::Int)
    println("Tableau de l'itération $iteration:")
    display(M) 
end

# Fonction pour le pivot
function pivot!(M::Matrix, i::Int, j::Int)
    m, n = size(M)
    @assert M[i, j] != 0
    M[i, :] = M[i, :] / M[i, j]
    for k in setdiff(1:m, i)
        M[k, :] -= M[k, j] * M[i, :]
    end
    return M
end

# Fonction pour obtenir les coûts réduits
function getReducedCosts(M::Matrix)
    m, n = size(M)
    return M[end, 1:n - 1]
end

# Fonction pour obtenir xB
function getxB(M::Matrix)
    m, n = size(M)
    return M[1:m - 1, end]
end

# Fonction pour la variable entrante
function enteringVar(M::Matrix)
    rc = getReducedCosts(M)
    index = argmin(rc)
    return rc[index] >= 0 ? -1 : index
end

# Fonction pour l'index de la variable sortante
function exitingVarIndex(M::Matrix{T}, enteringVar::Int) where T
    col = M[1:end-1, enteringVar]
    xB = getxB(M)
    m, n = size(M)
    index = -1
    val = T(Inf)
    for i in 1:m-1
        if (col[i] > 0) && (xB[i]/col[i] < val)
            val = xB[i]/col[i]
            index = i
        end
    end
    return index
end

# Fonction pour vérifier si un vecteur est one-hot
function isOneHot(v::Vector)
    n = length(v)
    return (sum(iszero, v) == n - 1) && (sum(isone, v) == 1)
end

# Fonction pour vérifier si la solution est optimale
function isoptimal(M)
    return enteringVar(M) == -1
end

# Fonction pour trouver la base initiale
function findInitialBasis!(M::Matrix)
    m, n = size(M)
    m -= 1
    n -= 1
    basis = [-1 for _ in 1:m]
    for i in 1:n
        if isOneHot(M[1:end - 1, i])
            index = findfirst(isone, M[:, i])
            basis[index] = i
        end
    end
    @assert !any(t -> t == -1, basis) "Problème non canonique"
    for i in 1:m
        j = basis[i]
        pivot!(M, i, j)
    end
    return basis
end

# Fonction pour le solveur du simplexe
function simplexSolver(A::Matrix{T}, b::Vector{T}, c::Vector{T}, verbose::Bool = false) where T
    M = [A b; c' 0]
    basis = findInitialBasis!(M)
    k = 1
    nmax = 1000
    while !isoptimal(M) && k < nmax
        k += 1
        verbose && display(M)
        entering = enteringVar(M)
        entering == -1 && (println("------------"); return zeros(T, 1))
        leaving = exitingVarIndex(M, entering)
        leaving == -1 && (println("------------"); return zeros(T, 1))
        verbose && @show (entering, leaving)
        basis[leaving] = entering
        pivot!(M, leaving, entering)
    end
    verbose && display(M)
    m, n = size(M)
    xstar = zeros(T, n - 1)
    xstar[basis] = getxB(M)
    return xstar
end
```

```julia
#feasible prend un tableau du simplex et retourne si il contient une solution primal rélisable
function feasible(M::Matrix{T}) where T
    xb = getxB(M)
    return all(xb .>= 0)
end

#retourrne la ligne du tableau du simplex dont la variable de base va sortir de la base
function findLeavingVarInDual(M::Matrix{T}) where T
    xb = getxB(M)
    return argmin(xb)
end

#retourne la variable qui va entrer dans la base. Si aucune variable ne peut entrer dans
#la base, on retourne -1.
function findEnteringVarInDual(M::Matrix{T}, leaving::Int) where T
    col = M[leaving, 1:end-1]
    reduced_costs = getReducedCosts(M)
    min_cost = T(Inf)
    entering_var = -1
    for i in 1:length(col)
        if col[i] < 0 && -reduced_costs[i] / col[i] < min_cost
            min_cost = -reduced_costs[i] / col[i]
            entering_var = i
        end
    end
    return entering_var
end

function DualsimplexSolver(A::Matrix{T}, b::Vector, c::Vector; 
        verbose::Bool = false) where T
    @assert all(c .>= 0) #dual feasibility
    M = [A b; c' 0]
    basis = findInitialBasis!(M)
    k = 1
    nmax = 1000
    while !feasible(M) && k < nmax
        k+=1
        verbose && display(M)
        leaving = findLeavingVarInDual(M)
        entering = findEnteringVarInDual(M, leaving)
        if entering == -1
            println("aucune variable ne peut entrer dans la base.")
            println("problème non réalisable")
        end
        verbose && @show (entering, leaving)
        basis[leaving] = entering
        pivot!(M, leaving, entering)
    end
    verbose && display(M)
    m, n = size(M)
    xstar = zeros(T, n - 1)
    xstar[basis] = getxB(M)
    return xstar
end
```
